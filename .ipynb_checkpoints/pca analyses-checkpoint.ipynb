```python
#Importing necessary libraries for data manipulation, visualization, and feature scaling.
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
#Reading a csv file and storing it in a DataFrame, header=None means that the csv file doesn't have column names.
df=pd.read_csv('emotions.csv') 
```

```python
#Filtering the rows of the dataframe where last column's value is 'NEUTRAL'.
df=df[df.iloc[:,-1]!='NEUTRAL']
```

```python
df.iloc[:,-1]=[0 if i=="NEGATIVE" else 1 for i in df.iloc[:,-1].values ]
```

```python
df
```

```python
def drop_outliers_MAD(df, columns=None, threshold=20):
    """
    Drops the outlier rows from a DataFrame using the MAD method
    
    Parameters:
    - df: DataFrame to remove outliers from
    - columns: List of columns to check for outliers (default: None, checks all columns)
    - threshold: threshold multiplier for MAD (default: 3.5)
    
    Returns:
    - DataFrame with outliers removed
    """
    if columns is None:
        columns = df.columns
    for col in columns:
        median = df[col].median()
        mad = df[col].mad()
        lower_bound = median - (threshold * mad)
        upper_bound = median + (threshold * mad)
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df
```

```python
dd=drop_outliers_MAD(df)
```

```python
dd
```

```python
X=dd.iloc[:,:-1].values
y= dd.iloc[:,-1].values 
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
#Splitting the data into training and testing sets using the train_test_split function from sklearn.model_selection.
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
from sklearn.naive_bayes import GaussianNB
# Train the model
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print(acc)
```

# PCA 


```python
from sklearn.decomposition import PCA
import numpy as np
```

```python
# Create an instance of PCA
pca = PCA(n_components=2)

# Fit the PCA model to the data
pca.fit(X)

# Apply dimensionality reduction to the data
X_pca = pca.transform(X)

# Print the explained variance ratio
print(pca.explained_variance_ratio_)

# Print the principal components
print(pca.components_)
```

```python
# Plot the data in the new 2-dimensional space using different colors for different classes
plt.scatter(X_pca[y==0, 0], X_pca[y==0, 1], c='blue', label='Class 0')
plt.scatter(X_pca[y==1, 0], X_pca[y==1, 1], c='red', label='Class 1')


# Add x-label, y-label and title
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA Plot')
# Add a legend
plt.legend(title='Classes')
# Show the plot
plt.show()
```

the two classes N and P, when projected to a two-dimensional space, can be separable up to some extent. Other observations can be that the N class is spread incede the P class.

```python
#Splitting the data into training and testing sets using the train_test_split function from sklearn.model_selection.
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    X_pca, y, test_size=0.2, random_state=1, stratify=y)
```

```python
from sklearn.naive_bayes import GaussianNB
# Train the model
clf = GaussianNB()
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
print(acc)
```

```python
def calculatacc(X,y):
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=1, stratify=y)
    clf = GaussianNB()
    clf.fit(X_train, y_train)
    y_pred=clf.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    print(acc)
    return acc
    
```

```python
accs=[]
for i in range(10):
    pca = PCA(n_components=i+1)

    # Fit the PCA model to the data
    pca.fit(X)

    # Apply dimensionality reduction to the data
    X_pca = pca.transform(X)
    accs.append(calculatacc(X_pca,y))
```

```python
plt.plot(accs)
plt.show()
```
